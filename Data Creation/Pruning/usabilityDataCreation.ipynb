# Test to see if i can Automate the usability of a video

```python
import os
import cv2
import shutil
import pandas as pd
import numpy as np
```

# Load in video Titles

```python
saveLocation = "D:/Thesis/Video-Auslan-sorted"
# on a scale of 1 to 4
usability = {}
for word in os.listdir(saveLocation):
    wordLocation = saveLocation + '/' + word
    # Loop through every video
    for videoName in os.listdir(wordLocation): 
        usability[videoName] = None
```

# Load in useability scores

```python
usability = {}
f = open("useability.txt", "r")
line = f.readline()
while (line):
    try:
        if line.split("_|_")[1] == 'None\n':
            usability[line.split("_|_")[0]] = None
        else:
            usability[line.split("_|_")[0]] = int(line.split("_|_")[1])
    except:
        if line != '\n':
            print(line)
    line = f.readline()
f.close()
usability
```

# Load in Video Lengths 

```python
videoLength = {}
f = open("videoLength.txt", "r")
line = f.readline()
while (line):
    try:
        if line.split("_|_")[1] == 'None\n':
            videoLength[line.split("_|_")[0]] = None
        else:
            videoLength[line.split("_|_")[0]] = int(line.split("_|_")[1])
    except:
        if line != '\n':
            print(line)
    line = f.readline()
f.close()
videoLength
```

# Find length of all videos

```python
saveLocation = "D:/Thesis/Video-Auslan-sorted"
videoLength = {}
for word in os.listdir(saveLocation):
    wordLocation = saveLocation + '/' + word
    # Loop through every video
    for videoName in os.listdir(wordLocation): 
        # Load video Files
        videoLocation = wordLocation + '/' + videoName
        frames = []
        cap = cv2.VideoCapture(videoLocation)
        numFrames = 0
        while cap.isOpened():
            ret, _ = cap.read()
            if not ret:
                break
            numFrames += 1
        videoLength[videoName] = numFrames
        cap.release()
videoLength
```

# Save Length of all videos into videoLength.txt

```python
videoLength
f = open("videoLength.txt", "w")
for video in videoLength:
    if videoLength[video] is None:
        f.write(video + "_|_" + 'None' + '\n')
    else:
        f.write(video + "_|_" + str(videoLength[video]) + '\n')
f.close()
videoLength
```

# Make All corrupted videos a usability of 0

```python
corrupted = ['AJBB1c2b',
'AJPB1c2b',
'AMMA1c2a',
'AWGB1c2a',
'BAOBB1c2a',
'BCPA1c2a',
'BDLA1c2a',
'BFPB1c2a',
'BFSA1c2a',
'BGLB1c2b',
'BRCA1c2a',
'BRVA1c2b',
'BSSA1c2b',
'BCHA2c7a',
'SSNA1c2a']
```

```python
for key in usability.keys():
    for corruptName in corrupted:
        if corruptName in key:
            usability[key] = 0
        elif videoLength[key] == 0:
            usability[key] = 0
```

# Calculate average length of video Based off usability scale

```python
useabilityArrayDic = {i:[] for i in range(0,5)}
useTotalLength = {i:0 for i in range(0,5)}
numTotalLength = {i:0 for i in range(0,5)}
for key in usability.keys():
    if videoLength[key] is not None and usability[key] is not None:
        useabilityArrayDic[int(usability[key])].append((key,videoLength[key]))
        numTotalLength[int(usability[key])] += 1
        useTotalLength[int(usability[key])] += videoLength[key]
meanTotalLength = {i:useTotalLength[i]/numTotalLength[i] for i in range(1,5)}
meanTotalLength
```

# Calculate average length of video based off video set

```python
videoSets = set()
videoDict = {}
for i in usability.keys():
    videoSets.add(str(i).split('_')[1])
    videoDict[i] = str(i).split('_')[1]
```

```python
useabilityArrayDic = {i:[] for i in videoSets}
useTotalLength = {i:0 for i in videoSets}
numTotalLength = {i:0 for i in videoSets}
for key in usability.keys():
    if videoLength[key] is not None:
        useabilityArrayDic[videoDict[key]].append((key,videoLength[key]))
        numTotalLength[videoDict[key]] += 1
        useTotalLength[videoDict[key]] += videoLength[key]
    else:
        print(key)
meanTotalLength = {}
for key in videoSets:
    meanTotalLength[key] = useTotalLength[key]/numTotalLength[key]
meanTotalLength
```

# Create Predict Data

```python
saveLocation = "D:/Thesis/Video-Auslan-sorted"
dataFrame = pd.DataFrame
videoTitle = []
videoLocation = []
videoLengthArr = []
meanLengthArr = []
for folder in os.listdir(saveLocation):
    for video in os.listdir(saveLocation + '/' + folder):
        if usability[video] is None:
            videoTitle.append(video)
            videoLocation.append(saveLocation + '/' + folder + '/' + video)
            videoLengthArr.append(videoLength[video])
            meanLengthArr.append(meanTotalLength[videoDict[video]])
dataFrame = pd.DataFrame(np.array([videoTitle, videoLocation, videoLengthArr, meanLengthArr]).T)
dataFrame.to_csv('PathUsababilityPredictData.csv')
```

```python
saveLocation = "D:/Thesis/Video-Auslan-sorted"
dataFrame = pd.DataFrame
videoTitle = []
videoLocation = []
videoLengthArr = []
meanLengthArr = []
usabilityArr = []
for folder in os.listdir(saveLocation):
    for video in os.listdir(saveLocation + '/' + folder):
        if usability[video]:
            videoTitle.append(video)
            videoLocation.append(saveLocation + '/' + folder + '/' + video)
            videoLengthArr.append(videoLength[video])
            meanLengthArr.append(meanTotalLength[videoDict[video]])
            usabilityArr.append(usability[video])
dataFrame = pd.DataFrame(np.array([videoTitle, videoLocation, videoLengthArr, meanLengthArr, usabilityArr]).T)
dataFrame.to_csv('PathUsababilityTrainData.csv')
```
