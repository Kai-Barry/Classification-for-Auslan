```python
numClassification = {}
for value in get_path_movement().values():
    if value in numClassification:
        numClassification[value] += 1
    else:
        numClassification[value] = 1
print(numClassification)
labels = [key for key in numClassification.keys() if key is not None and key != 'Unknown\n']
values = [value for key, value in numClassification.items() if key is not None and key != 'Unknown\n']
plot_bar(labels, values, 'classification', 'number of classification', 'Movement classifcation')
```
