```python
def is_comprised_of_characters(input_string, allowed_characters):
    for char in input_string:
        if char not in allowed_characters:
            return False
    return True
```

```python
import string

lowercase_characters = list(string.ascii_lowercase)
uppercase_characters = list(string.ascii_uppercase)
numWords = 0
otherWords = set()
onlyNumWords = {}
for word in allWords:
    if is_comprised_of_characters(word, lowercase_characters + uppercase_characters + ['-','(', ')', '1', '2','3','4']):
        numWords += 1
        onlyNumWords[word] = numAllWords[word]
    else:
        otherWords.add(word)
numWords
```

```python
import cv2
print(cv2.__version__)
import os
import numpy as np
signScore = {}
print(range(len(auto['test']['age'])))
print(len(auto['test']['age']['names']))
print(len(verified['test']['age']['names']))
print(auto['test']['age']['global_times'])
location = 'D:/BOBSL/BOBSL200G/bobsl/videos/'
playbackSpeed = int((1/25)*1000)

for i in range(25):
    videoName = auto['test']['age']['names'][i]
    videoLocation = location + videoName + '.mp4'
    if os.path.exists(videoLocation):
        print("Video: ", videoLocation, "exists")
        video_capture = cv2.VideoCapture(videoLocation)
        fps = video_capture.get(cv2.CAP_PROP_FPS)
        start = int(auto['test']['age']['global_times'][i] * fps)
        end = start + 20
        numFrame = 0
        videoFrames = []
        print("Caputuring frames")
        while True:
            ret, frame = video_capture.read()
            if not ret or numFrame > end:
                break
            if numFrame < start:
                numFrame += 1
                continue
            videoFrames.append(frame)
            numFrame += 1
        video_capture.release()
        print('Frames Caputred')
        if len(videoFrames) > 0:
            frameNum = 0
            while True:
                if frameNum >= len(videoFrames):
                    frameNum = 0
                try:
                    cv2.imshow('MediaPipe Pose', videoFrames[frameNum])
                except:
                    print(videoName, frameNum, len(videoFrames))
                    break
                frameNum += 1
                k = cv2.waitKey(playbackSpeed) & 0xFF
                if k == ord('q'):
                    break
                elif k == ord('p'):
                    signScore[videoName] = True
                elif k == ord('l'):
                    signScore[videoName] = False
            cv2.destroyAllWindows() 
        cv2.destroyAllWindows()
    else:
        print(videoName, 'does not exist')
    cv2.destroyAllWindows() 


```

```python
usability = {}

f = open("useability.txt", "r")
line = f.readline()
while (line):
    try:
        if line.split("_|_")[1] == 'None\n':
            usability[line.split("_|_")[0]] = None
        else:
            usability[line.split("_|_")[0]] = int(line.split("_|_")[1])
    except:
        if line != '\n':
            print(line)
    line = f.readline()
f.close()
usability
```
